```python
#Top 10 drinks with highest calcium
df.groupby(['Beverage_category','Beverage','Beverage_prep']).max()['Calories'].sort_values(ascending=False)[:10]
```

```python
df.groupby(['Beverage_category','Beverage', 'Beverage_prep']).max()['Calories'].sort_values(ascending=False)[:10].plot(kind='barh', color='indigo')
plt.xlabel("Calories in the Drink")
plt.ylabel("Drink")
plt.title("Top 10 Drinks with High value of Calcium");
```

```python
#Top 10 drinks with highest Protein
df.groupby(['Beverage_category','Beverage','Beverage_prep']).max()['Protein (g) '].sort_values(ascending=False)[:10]
```

```python
df.groupby(['Beverage_category','Beverage','Beverage_prep']).max()['Protein (g) '].sort_values(ascending=False)[:10].plot(kind='barh',color='g')
```

```python
#Top 10 drinks with high value of sugar
df.groupby(['Beverage_category','Beverage','Beverage_prep']).max()['Sugars (g)'].sort_values(ascending=False)[:10]
```

```python
#Top 10 drinks with high value of sugar
df.groupby(['Beverage_category','Beverage','Beverage_prep']).max()['Sugars (g)'].sort_values(ascending=False)[:10].plot(kind='barh',color='purple')
```

```python
#Top 10 drinks with high Trans fat
df.groupby(['Beverage_category','Beverage','Beverage_prep']).max()['Trans Fat (g) '].sort_values(ascending=False)[:10]
```

```python
df.groupby(['Beverage_category','Beverage','Beverage_prep']).max()['Trans Fat (g) '].sort_values(ascending=False)[:10].plot(kind='barh',color='orange')
plt.xlabel('Trans fat')
plt.ylabel('Drink')
```
